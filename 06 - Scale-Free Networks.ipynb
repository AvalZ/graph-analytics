# Power Law generative model

## Barabasi-Albert model

Each new node enters the network with $m$ links and it connects with another node with a probability depending to the degree $k_i$ of node $i$.

$$ \Pi(k_i) = \frac{k_i}{\sum_{j} k_j} $$

This distribution creates a *riches get richer* situation.

Each node enters the network with at least $m$ links, but this is a *power law* distribution, so you could have a node with a higher number of nodes.

## First Mover Advantage

The *older* I am, the *stronger* I am in the network.

### Bianconi-Barabasi model - Intrinsic quality

In 1997, Google was the "late Web arrival", but in 1997-2000 it became the most significant node in the network.

This is not described by the model, we need a notion of *intrinsic quality* of a node.

This could also be described by a **fitness** value.

The probability that a link of a new node connects to a pre-existing node $i$ is proportional to both the $i$'s degree and fitness.

$$ \Sigma(k_i) = \frac{k_i \cdot \eta_i}{\sum_{j} k_j \cdot \eta_j } $$

### Physics intuition

Some networks can undergo *Bose-Einstein condensation*.

#### Fitness $\rightarrow$ Energy

Each node with fitness $\eta_j$ has energy $\varepsilon_j$.

Complex networks with preferential attachments shows three different growth models:

* **first mover advantage** (scale free)
* **fit gets rich** (scale free)
* **winner takes all** (star-like hub-and-spoke)
